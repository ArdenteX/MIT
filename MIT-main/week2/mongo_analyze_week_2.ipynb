```python
from pyecharts import options as opts
from pyecharts.charts import Map, Pie, Bar
from pyecharts.datasets import register_url
from analyze_week_2 import PlotMap, PlotPie, PlotBar
```

```python
register_url("https://echarts-maps.github.io/echarts-countries-js/")
```

```python
p = PlotMap()
max_data, min_data, count_list = p.convert_data()
```

```python
count_list
```

```python

us = (
    Map()
    .add("Covid-19 Case", count_list, maptype="美国", is_roam=True, emphasis_label_opts={
        'label':{
            'show': True
        }
    })

    .set_global_opts(
        title_opts={
             'text': 'USA Covid-19 Case',
             'subtext': 'Data from 	CDC',
             'left': 'right'
        },
        visualmap_opts={
             'left': 'right',
              'min': min_data,
              'max': max_data,
              'inRange': {
              'color': [
                  '#313695',
                  '#4575b4',
                  '#74add1',
                  '#abd9e9',
                  '#e0f3f8',
                  '#ffffbf',
                  '#fee090',
                  '#fdae61',
                  '#f46d43',
                  '#d73027',
                  '#a50026'
                ]
              },
              'text': ['High', 'Low'],
              'calculable': True
        },
        tooltip_opts={
            'trigger': 'item',
            'showDelay': 0,
            'transitionDuration': 0.2
        },
        toolbox_opts={
            'show': True,
            'left': 'left',
            'top': 'top',
            'feature': {
            'dataView': { 'readOnly': 'false' },
            'restore': {},
            'saveAsImage': {}
            }
        },
    )
    
)
```

```python
us.render_notebook()
```

```python

max_d, min_d, pie = PlotPie().convert_data()
```

```python
e = (
    Pie(init_opts=opts.InitOpts(bg_color='#2c343c', animation_opts=opts.AnimationOpts(animation_easing='elasticOut')))
    .add("USA", pie,
         radius='55%',
         center=['50%', '50%'],
         rosetype='radius',
         label_opts={
             'color': 'rgba(255, 255, 255, 0.3)',
             'smooth': 0.2,
             'length': 10,
             'length2': 20
         },
    )
    .set_series_opts(
        linestyle_opts={
            'color': 'rgba(255, 255, 255, 0.3)',

        },
        itemstyle_opts={
            'color': '#c23531',
            'shadowBlur': 200,
            'shadowColor': 'rgba(0, 0, 0, 0.5)',
        },


    )
    .set_global_opts(
        title_opts={
            'text': 'Age Groups',
            'left': 'center',
            'top': 20,
            'textStyle': {
            'color': '#ccc'
            }
        },
        tooltip_opts={
            'trigger': 'item'
        },
        visualmap_opts={
             'show': False,
             'min': min_d,
             'max': max_d,
             'inRange': {
                'colorLightness': [0.3, 0.6]
              }
        },
    )
   
)
```

```python
e.load_javascript()
```

```python
e.render_notebook()
```
